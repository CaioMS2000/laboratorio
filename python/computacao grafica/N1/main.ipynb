```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = 4
ys = 4
zs = 4
```

```python

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])
```

```python
fig.savefig('dot3D.png')
plt.show()
```
